```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 100)
Y = np.sqrt(X)
Y1 = np.sin(X)
Y2 = np.cos(X)
```

# Inisialisasi Figure and Axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(X, Y)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(X, Y1)
ax.plot(X, Y2)
```

### shortcut

```python
plt.figure(figsize=(6,6))
plt.plot(X, Y1, 'r--', linewidth=0.5)
plt.plot(X, Y2, 'bp', markersize=3)
plt.xlim(0, 10)
plt.ylim(-2, 2)
plt.axis('equal')
plt.title('sin and cos')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(X, Y1, 'r--', linewidth=0.5)
ax.plot(X, Y2, 'bp', markersize=3)
ax.set(xlim=(0, 10), ylim=(-2, 2), title=('sin dan cos'), xlabel='x', ylabel='y')
ax.axis('equal')
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(X, Y1, 'r--', linewidth=0.5, label='sin')
ax.plot(X, Y2, 'bp', markersize=3, label='cos')
ax.set(xlim=(0, 10), ylim=(-2, 2), title=('sin dan cos'), xlabel='x', ylabel='y')
ax.axis('equal')
ax.legend();
```

# Subplot

### Matlab Style

```python
plt.figure(figsize=(15, 10))

plt.subplot(231)
plt.hist(numbers, bins=80, density=1);

plt.subplot(233)
plt.hist(numbers, bins=80, density=1);

plt.subplot(235)
plt.scatter(x, y, c=rad , s=rad*90, cmap='seismic')
plt.axis('equal');
```

### Matplotlib Style

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 10))

ax[0, 0].hist(numbers, bins=80, density=1);

ax[0, 2].hist(numbers, bins=80, density=1);

ax[1, 1].scatter(x, y, c=rad , s=rad*90, cmap='seismic')
ax[1, 1].axis('equal');
```

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 10), sharex=True, sharey=True)

ax[0, 0].hist(numbers, bins=80, density=1);

ax[0, 2].hist(numbers, bins=80, density=1);

ax[1, 1].scatter(x, y, c=rad , s=rad*90, cmap='seismic');
```

# Teks and Anotation

```python
numbers = np.random.randn(10000)
```

```python
plt.figure(figsize=(6, 6))
plt.hist(numbers, bins=80, density=1);
plt.ylim(0, 1)
plt.axvline(0, c='r');
plt.text(0, 0.8, 'mean= 0,', ha='center')
plt.annotate('mean= 0,', xy=(0, 0.4), xytext=(2, 0.6), ha='center', arrowprops={'arrowstyle': '-|>'})
```

# 3D Plot

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
```

```python
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)

x, y = np.meshgrid(x, y)
z = np.sqrt(x**2 + y**2 + 10)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, antialiased=False, cmap='winter');
ax.set(xlabel='x', ylabel='y', zlabel='z')
ax.view_init(elev=10)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, antialiased=False, cmap='winter');
ax.set(xlabel='x', ylabel='y', zlabel='z')
ax.view_init(elev=-30)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, antialiased=False, cmap='winter');
ax.set(xlabel='x', ylabel='y', zlabel='z')
ax.view_init(elev=-30, azim=45)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_wireframe(x, y, z, cmap='winter');
ax.set(xlabel='x', ylabel='y', zlabel='z')
ax.view_init(elev=-30)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.contour3D(x, y, z, 50, cmap='winter');
ax.set(xlabel='x', ylabel='y', zlabel='z')
ax.view_init(elev=-30, azim=45)
```
