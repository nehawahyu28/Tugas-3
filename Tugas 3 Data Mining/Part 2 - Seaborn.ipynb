```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('linear.csv')
```

```python
df.head()
```

```python
df1 = df[df.tipe== 'data1']
df2 = df[df.tipe== 'data2']
plt.scatter(df1.X, df1.y)
plt.scatter(df2.X, df2.y)
```

### Seaborn

```python
plt.figure(figsize=(6, 6))
sns.scatterplot(x= "X", y= "y", data=df, hue="tipe")
```

```python
plt.figure(figsize=(6, 6));
sns.lmplot(x="X", y="y", data=df, hue="tipe");
```

```python
plt.figure(figsize=(6, 6));
sns.distplot(df.y, bins=20);
```

# Seaborn and Pandas Dataframe

### Statistical Summary

```python
df = pd.read_csv('cereal.csv')
```

```python
df.head()
```

### Missing Value

```python
plt.figure(figsize=(6, 6));
sns.heatmap(~df.isna(), cmap='Blues', cbar=False, );
```

### Correlation Heatmap

```python
plt.figure(figsize=(6, 6));
sns.heatmap(df.corr(), cmap='bwr', cbar=False, vmin=-1, vmax=1, annot=True, fmt='.1f')
```

### Pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='shelf', vars=['calories', 'fat', 'rating'])
```

```python
sns.pairplot(df, hue='shelf', vars=['calories', 'fat', 'rating'], kind='reg')
```

```python
sns.jointplot(x='potass', y='fiber', data=df, color='r', marginal_kws={'bins': 25}, kind='reg')
```

# Categorical Data

### Persebaran

```python
sns.catplot(x='mfr', y='rating', data=df)
```

```python
sns.boxplot(x='mfr', y='rating', data=df)
```

```python
sns.violinplot(x='mfr', y='rating', data=df)
```

### Count

```python
sns.catplot(x='mfr', data=df, kind='count')
```

```python
sns.countplot(x='mfr', data=df, hue='shelf')
```
